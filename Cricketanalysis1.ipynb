Matches played by the teams in Asia Cup.

```python
x= df['Team'].value_counts()
y= df['Team'].unique()
ex=[0,0,0,0,0.1,0.2,0.2]

plt.pie(x, labels=y,autopct="%0.0f%%",explode=ex)
plt.title("Matches played by teams")
plt.show()
```
